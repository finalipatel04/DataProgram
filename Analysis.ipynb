```python
def rating(rating):
    if rating>=7.0:
        return "Block-Buster"
    if rating>=6.0:
        return "Hit"
    else:
        return "Average or Flop"
```

```python
data['Rating_cat']=data['Rating'].apply(rating)
```

```python
list1=[]
for value in data['Genres']:
    list1.append(value.split(','))
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
